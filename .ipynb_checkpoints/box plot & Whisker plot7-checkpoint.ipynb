# box plot & Whisker plot

```python
import matplotlib.pyplot as plt
x=[10,20,30,40,50,60,70]
plt.boxplot(x)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[10,20,30,40,50,60,70]

#notch
plt.boxplot(x,notch=True)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[10,20,30,40,50,60,70]

#horizontal box plot
plt.boxplot(x,vert=False)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[10,20,30,40,50,60,70]
# changing width
plt.boxplot(x,widths=0.2,labels=["Python"],patch_artist=True)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[10,20,30,40,50,60,70]
# changing width
plt.boxplot(x,widths=0.2,labels=["Python"],patch_artist=False,showmeans=True)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
# showing the outler value
x=[10,20,30,40,50,60,70,120]

plt.boxplot(x,widths=0.1,label=["Python"],patch_artist=False,showmeans=True)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[10,20,30,40,50,60,70,120]

#outler ko join karne ke liye hum whis parameter ka use karte hai.
plt.boxplot(x,widths=0.1,label=["Python"],patch_artist=False,showmeans=True,whis=2)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[10,20,30,40,50,60,70,120]

#changing the color of outler and his mark. Agar hum outler ko hide karna chahte hai to hum (sym=" ") ko khali chhod denge.
plt.boxplot(x,widths=0.1,label=["Python"],patch_artist=False,showmeans=True,sym="g+")
plt.legend()
plt.show()
```

```python
#changing the box property
#box ki property ko change karne ke liye boxprops prameter ka use karte hai. isme hum value as a dictionary pass karte hai.
import matplotlib.pyplot as plt
x=[10,20,30,40,50,60,70,120]

plt.boxplot(x,widths=0.1,label=["Python"],patch_artist=False,showmeans=True,sym="g+",
           boxprops=dict(color="r"))
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[10,20,30,40,50,60,70]

#hum yaha per cap ki value ko change bhi kar sakte hai. capprops parameter ki help se .
plt.boxplot(x,widths=0.1,label=["Python"],patch_artist=False,showmeans=True,sym="g+",capprops=dict(color="r"))
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[10,20,30,40,50,60,70,120]

#hum yaha per whisker ki value ko change bhi kar sakte hai. whiskerprops parameter ki help se .
plt.boxplot(x,widths=0.1,label=["Python"],patch_artist=False,showmeans=True,sym="g+",capprops=dict(color="r"),whiskerprops=dict(color="r"),
            flierprops=dict(markeredgecolor="blue"))
plt.legend()
plt.show()
```

```python
#multiple box plot
import matplotlib.pyplot as plt
x=[10,20,30,40,50,60,70]
x1=[10,20,30,40,50,60,90]

plt.boxplot(x,widths=0.1,label=["Python"],patch_artist=False,showmeans=True,sym="g+",capprops=dict(color="r"),whiskerprops=dict(color="r"))
plt.boxplot(x1,widths=0.1,label=["Python"],patch_artist=False,showmeans=True,sym="g+",capprops=dict(color="r"),whiskerprops=dict(color="r"))

plt.legend()
plt.show()
```

```python
#multiple box plot
import matplotlib.pyplot as plt
x=[10,20,30,40,50,60,70]
x1=[10,20,30,40,50,60,90]

y=[x,x1]
plt.boxplot(y,labels=["python","c++"],patch_artist=False,showmeans=True,sym="g+",capprops=dict(color="r"),whiskerprops=dict(color="r"))

plt.legend()
plt.show()
```
