# Matplotlib pie plot

```python
import matplotlib.pyplot as plt
X=[10,20,30,40]
y=["c","c++","java","python"]
plt.pie(X,labels=y)
plt.show()
```

```python
import matplotlib.pyplot as plt
X=[10,20,30,40]
y=["c","c++","java","python"]

ex=[0.3,0.0,0.0,0.0]
#exlode ka use hum , hamare parameter ko bada karne ke liye lagate hai.
plt.pie(X,labels=y,explode=ex)
plt.show()
```

```python
import matplotlib.pyplot as plt
X=[10,20,30,40]
y=["c","c++","java","python"]

ex=[0.3,0.0,0.0,0.0]
c=["r","g","y","b"]
plt.pie(X,labels=y,explode=ex,colors=c)
plt.show()
```

```python
import matplotlib.pyplot as plt
X=[10,20,30,40]
y=["c","c++","java","python"]

ex=[0.3,0.0,0.0,0.0]
c=["r","g","y","b"]

#printing the percentage(autopct="%0.1f%%")
plt.pie(X,labels=y,explode=ex,colors=c,autopct="%0.2f%%")
plt.show()
```

```python
import matplotlib.pyplot as plt
X=[10,20,30,40]
y=["c","c++","java","python"]

ex=[0.3,0.0,0.0,0.0]
c=["r","g","y","b"]

#using shadow , radius and  labeldistance
plt.pie(X,labels=y,explode=ex,colors=c,autopct="%0.2f%%",shadow=True,radius=0.8,labeldistance=0.8)
plt.show()
```

```python
import matplotlib.pyplot as plt
X=[10,20,30,40]
y=["c","c++","java","python"]

ex=[0.3,0.0,0.0,0.0]
c=["r","g","y","b"]

# changing the start value
plt.pie(X,labels=y,explode=ex,colors=c,autopct="%0.2f%%",shadow=True,radius=0.8,startangle=180)
plt.show()
```

```python
import matplotlib.pyplot as plt
X=[10,20,30,40]
y=["c","c++","java","python"]

ex=[0.3,0.0,0.0,0.0]
c=["r","g","y","b"]

#hum yaha per java,c,c++,python etc. ke fontside ko bhi badha sakte hai.uske liye hame textprops name ki property ka use karte hai.
plt.pie(X,labels=y,explode=ex,colors=c,autopct="%0.2f%%",shadow=True,radius=0.8,textprops={"fontsize":15})
plt.show()
```

```python
import matplotlib.pyplot as plt
X=[10,20,30,40]
y=["c","c++","java","python"]

ex=[0.3,0.0,0.0,0.0]
c=["r","g","y","b"]

#changing the direction
plt.pie(X,labels=y,explode=ex,colors=c,autopct="%0.2f%%",shadow=True,radius=0.8,textprops={"fontsize":15},counterclock=False)
plt.show()
```

```python
import matplotlib.pyplot as plt
X=[10,20,30,40]
y=["c","c++","java","python"]

ex=[0.1,0.0,0.0,0.0]
c=["r","g","y","b"]

#changing the edge property by using the wedgeprops parameter.
#Note: hum yaha per jo bhi property denge vo ek dictionart ke through denge.
plt.pie(X,labels=y,explode=ex,colors=c,autopct="%0.2f%%",shadow=True,radius=0.8,textprops={"fontsize":15},wedgeprops={"linewidth":1.5,"edgecolor":"black"})
plt.show()
```

```python
import matplotlib.pyplot as plt
X=[10,20,30,40]
y=["c","c++","java","python"]

ex=[0.3,0.0,0.0,0.0]
c=["r","g","y","b"]

#center property ko use karke hum center ko bhi change kar sakte hai.
plt.pie(X,labels=y,explode=ex,colors=c,autopct="%0.2f%%",shadow=True,radius=0.8,textprops={"fontsize":15},
        wedgeprops={"linewidth":1.5,"edgecolor":"black"},center=(9,9))
plt.show()
```

```python
import matplotlib.pyplot as plt
X=[10,20,30,40]
y=["c","c++","java","python"]

ex=[0.3,0.0,0.0,0.0]
c=["r","g","y","b"]

# hum yaha per label ko ratate bhi kar sakte hai.
plt.pie(X,labels=y,explode=ex,colors=c,autopct="%0.2f%%",shadow=True,radius=0.8,textprops={"fontsize":15},
        wedgeprops={"linewidth":1.5,"edgecolor":"black"},rotatelabels=True)
plt.show()
```

```python
import matplotlib.pyplot as plt
X=[10,20,30,40]
y=["c","c++","java","python"]

ex=[0.3,0.0,0.0,0.0]
c=["r","g","y","b"]

#using title , legend and changing the loction of legend
plt.pie(X,labels=y,explode=ex,colors=c,autopct="%0.2f%%",shadow=True,radius=0.8,textprops={"fontsize":15},
        wedgeprops={"linewidth":1.5,"edgecolor":"black"})
plt.title("Languages")
plt.legend(loc=3)
plt.show()
```

```python
# making a dot pie chart
plt.pie([1])
plt.show()
```

```python
#making multiple piechart
import matplotlib.pyplot as plt
X=[10,20,30,40]
X1=[40,30,20,10]
y=["c","c++","java","python"]


ex=[0.3,0.0,0.0,0.0]
c=["r","g","y","b"]

#using title , legend and changing the loction of legend
plt.pie(X,labels=y,radius=1.5)
plt.pie(X1,radius=0.5,colors=c)
plt.show()
```

```python
#making a white piechart Or ring piechart
import matplotlib.pyplot as plt
X=[10,20,30,40]
X1=[40,30,20,10]
y=["c","c++","java","python"]


ex=[0.3,0.0,0.0,0.0]
c=["r","g","y","b"]

#using title , legend and changing the loction of legend
plt.pie(X,labels=y,radius=1,colors=c)
plt.pie([1],radius=0.7,colors="w")
plt.show()
```

```python
#making a white piechart Or ring piechart (second way)
#making a white piechart Or ring piechart
import matplotlib.pyplot as plt
X=[10,20,30,40]
X1=[40,30,20,10]
y=["c","c++","java","python"]


ex=[0.3,0.0,0.0,0.0]
c=["r","g","y","b"]

#using title , legend and changing the loction of legend
plt.pie(X,labels=y,radius=1,colors=c)
cr=plt.Circle(xy=(0,0),radius=0.7,facecolor="w")
plt.gca().add_artist(cr)
plt.show()
```
