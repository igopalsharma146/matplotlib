# step Plot 

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[2,2,3,5,4]

#ms=marker size,mfc=marker face color
plt.step(x,y,marker="o",color="r",ms=10,mfc="g",label="python")

plt.title("python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.legend(loc=2)
plt.grid()
plt.show()
```
