# xticks(),yticks(),xlim(),ylim(),axis() in matplotlib python

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[3,1,5,2,4]

plt.plot(x,y)

plt.xticks(x,labels=["c","c++","python","java","js"])
plt.yticks(y)


plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[3,1,5,2,4]

plt.plot(x,y)

# plt.xticks(x,labels=["c","c++","python","java","js"])
# plt.yticks(y)

# hum yaha per xlim() or ylim() ka use karke x or y ki limit ko bhi badha sakte hai.
plt.xlim(0,10)
plt.ylim(0,10)
plt.show()
```
