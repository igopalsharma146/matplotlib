```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5,6]
y=[2,2,5,6,3,4]

#changing the line through linefmt (Solid line,Dashed line,Dash-dot line,Dotted line) parameter
plt.stem(x,y,linefmt=":")
plt.stem(x,y,linefmt="--")

#changing the marker and his color
plt.stem(x,y,markerfmt="r+")

#changing the bottom line
# plt.stem(x,y,bottom=4)

#changing the base line color
plt.stem(x,y,basefmt="g")

#using the label
plt.stem(x,y,label="Python")

#changing the color according to base color
# plt.stem(x,y,linefmt=":",label="Python",markerfmt="ro",basefmt="g",use_line_collection=False)

#we can also change here his orientation
# plt.stem(x,y,linefmt=":",label="Python",markerfmt="ro",basefmt="g",orientation="horizontal")

plt.legend()
plt.show()
```
