```python
#scatter plot
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
number=[4,5,7,9,4,5,8]
plt.scatter(day,number)
plt.title("scatter plot",fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.show()
```

```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
number=[4,5,7,9,4,5,8]

#changing the color
plt.scatter(day,number,color="r")
plt.title("scatter plot",fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.show()
```

```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
number=[4,5,7,9,4,5,8]

#changing the color
c=["r","y","g","b","r","g","r"]
plt.scatter(day,number,color=c)
plt.title("scatter plot",fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.show()
```

```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
number=[4,5,7,9,4,5,8]

#changing the color
c=["r","y","g","b","r","g","r"]

#changing the size
sizes=[10,16,7,14,8,9,12]
plt.scatter(day,number,color=c,s=sizes)
plt.title("scatter plot",fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.show()
```

```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
number=[4,5,7,9,4,5,8]

#changing the color
c=["r","y","g","b","r","g","r"]

#changing the size
sizes=[10,16,7,14,8,9,12]

#changing the transparency
plt.scatter(day,number,color=c,s=sizes,alpha=.5)
plt.title("scatter plot",fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.show()
```

```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
number=[4,5,7,9,4,5,8]

#changing the color
c=["r","y","g","b","r","g","r"]

#changing the size
sizes=[100,160,170,140,181,159,129]

#changing the marker and edgecolor
plt.scatter(day,number,color=c,s=sizes,marker="*",edgecolor="black",linewidth=.7)
plt.title("scatter plot",fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.show()
```

```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
number=[4,5,7,9,4,5,8]

sizes=[100,160,170,140,181,159,129]
#changing the color trough colormap
colors=[10,49,59,23,56,20,10]
plt.scatter(day,number,c=colors,cmap="viridis",s=sizes)
plt.title("scatter plot",fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.colorbar()
plt.show()
```

```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
number=[4,5,7,9,4,5,8]

sizes=[100,160,170,140,181,159,129]
#changing the color trough colormap
colors=[10,49,59,23,56,20,10]

#cmap=[Accent,Blues,BrBG,BuGn,BuPu ,etc....]
plt.scatter(day,number,c=colors,cmap="viridis",s=sizes)
plt.title("scatter plot",fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("No.",fontsize=15)

#label
t=plt.colorbar()
t.set_label("Colorbar")
plt.show()
```

```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
number=[4,5,7,9,4,5,8]
number2=[4,7,8,5,4,2,9]
sizes=[100,160,170,140,181,159,129]
#changing the color trough colormap
colors=[10,49,59,23,56,20,10]
plt.scatter(day,number,c=colors,cmap="viridis",s=sizes)
plt.scatter(day,number2,color="r",s=sizes)

plt.title("scatter plot",fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("No.",fontsize=15)

#label
t=plt.colorbar()
t.set_label("Colorbar")
plt.show()
```
