# matplotlib stack plot

```python
#stackplot
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
area=[3,5,2,7,4]
plt.stackplot(x,area)
```

```python
#multiple stackplot
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
area1=[3,4,2,7,4]
area2=[2,3,3,4,5]
area3=[4,3,5,3,2]
l=["area1","area2","area3"]
plt.stackplot(x,area1,area2,area3,labels=l,colors=["r","b","m"])
plt.legend()
plt.show()
```

```python
#multiple stackplot
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
area1=[3,4,2,7,4]
area2=[2,3,3,4,5]
area3=[4,3,5,3,2]
l=["area1","area2","area3"]

#base line
plt.stackplot(x,area1,area2,area3,labels=l,colors=["r","b","m"],baseline="sym")
plt.legend()
plt.show()
```

```python
#multiple stackplot
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
area1=[3,4,2,7,4]
area2=[2,3,3,4,5]
area3=[4,3,5,3,2]
l=["area1","area2","area3"]
plt.stackplot(x,area1,area2,area3,labels=l,colors=["r","b","m"])

plt.title("Python")
plt.xlabel("X-axis")
plt.ylabel("y-axis")
# plt.grid()

plt.legend()
plt.show()
```
