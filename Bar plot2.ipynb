```python
import matplotlib.pyplot as plt
x=["Python","c++","c","java"]
y=[90,70,80,92]
z=[55,32,45,42]
#giving the label and changing the fontsize
plt.xlabel("language",fontsize=10)
plt.ylabel("Marks out of 100.",fontsize=10)
plt.title("Bar Plot",fontsize=10)

#we can make here multiple bar graph
plt.bar(x,y,width=0.2,color="blue",edgecolor="black",linewidth=2,label="Popularity 1")
plt.bar(x,z,width=0.2,color="yellow",edgecolor="black",linewidth=2,label="Popularity 2")
#but here we can see that graph is doing overlap
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=["Python","c++","c","java"]
y=[90,70,80,92]
z=[55,32,45,42]

width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]

#giving the label and changing the fontsize
plt.xlabel("language",fontsize=10)
plt.ylabel("Marks out of 100.",fontsize=10)
plt.title("Bar Plot",fontsize=10)

#we can make here multiple bar graph
plt.bar(p,y,width,color="blue",edgecolor="black",linewidth=2,label="Popularity")
plt.bar(p1,z,width,color="yellow",edgecolor="black",linewidth=2,label="Popularity1")

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=["Python","c++","c","java"]
y=[90,70,80,92]
z=[55,32,45,42]

width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]

#giving the label and changing the fontsize
plt.xlabel("language",fontsize=10)
plt.ylabel("Marks out of 100.",fontsize=10)
plt.title("Bar Plot",fontsize=10)

#we can make here multiple bar graph
plt.bar(p,y,width,color="blue",edgecolor="black",linewidth=2,label="Popularity")
plt.bar(p1,z,width,color="yellow",edgecolor="black",linewidth=2,label="Popularity1")

#here , X axis per hamari language ke original name print nahi ho rahe hai..., 
# to hum yaha per uske original name xticks method ka use karke print kara sakte hai.
# plt.xticks(p+width,x)

#names ko center me karne ke liye ke liye hum width ko divide kar sakte hai. and hum rotation parameter ki help se unhe set bhi kar sakte hai
plt.xticks(p+width/2,x,rotation=70)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=["Python","c++","c","java"]
y=[90,70,80,92]
z=[55,32,45,42]

width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]

#giving the label and changing the fontsize
plt.xlabel("language",fontsize=10)
plt.ylabel("Marks out of 100.",fontsize=10)
plt.title("Bar Plot",fontsize=10)

#here we can also make the horizontal graph.
plt.barh(p,y,width,color="blue",edgecolor="black",linewidth=2,label="Popularity")
plt.barh(p1,z,width,color="yellow",edgecolor="black",linewidth=2,label="Popularity1")

#hum yaha per legend label ki position ko bhi set kar sakte hai,legend ke under location parameter ko use karke.
plt.yticks(p+width/2,x,rotation=70)
plt.legend()
plt.show()
```
